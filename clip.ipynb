```python
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

Эксперимент 1 (измененный lr)

```python
phrase = "На фотографии изображена собака, погруженная почти с головой в бассейн"
image = Image.open("image_1.png")
inputs = processor(text=[phrase], images=image, return_tensors="pt", padding=True)
```

```python
with torch.no_grad():
    outputs = model(**inputs)
    text_embed = outputs.text_embeds
    image_embed = outputs.image_embeds
```

```python
text_embed = text_embed / text_embed.norm(p=2, dim=-1, keepdim=True)
image_embed = image_embed / image_embed.norm(p=2, dim=-1, keepdim=True)
cosine_similarity = (text_embed * image_embed).sum(dim=-1).item()

print(f"CLIP-score для 1-го промпта про собаку: {cosine_similarity:.4f}")
```
